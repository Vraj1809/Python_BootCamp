```python
# ROUNDING THE NUMBER

value = 4.5


print(f"Floor of the value: {math.floor(value)}")

print(f"Ceil of the value: {math.ceil(value)}")

print(f"Round of the value: {round(value)}")

print(f"The Round of the 5.5: {round(5.5)}")

# now for 4.5 we got 54 and for 5.5. we got 6, that's because normally the value would made even while using the round function for ease

print(round(6.5))
print(round(7.5))
# however this would only happens when there isn .5 but if it is more than .5 then whether it is even or odd it will be converetd to the next number for rounding 

print(round(6.7))
print(round(7.7))
```

```python
# MATHEMATICAL CONSTANTS

from math import pi 

print(f"The value of constant pi: {pi}")

print(f"The value of constant e: {math.e}")

print(f"The value of constant tau: {math.tau}")

print(f"The value of constant infinity: {math.inf}")

print(f"The value of constant nan: {math.nan}")
```

```python
# LOGARITHMIC VALUES

#log with base e
print(f"The value of e with base e: {math.log(math.e)}")

print(f"The value of 10 with base e: {math.log(10)}")

print(f"Invalid value for log that is 0 {math.log(0)}")
```

```python
# log with custom base 
# math.llog(value, base)

print(f"The value of 10 with base 10: {math.log(10,10)}")
```

```python
# TRIGNOETRIC FUNCTIONS
# sine with answer in radian
print(math.sin(10))

#converting radians to degrees
print(math.degrees(pi/2))

#converting degrees to radians
print(math.radians(180))
```

RANDOM MODULE

```python
import random
# RANDOM NUMBER WITHOUT ANNY SEED - for differnt machines each random value would bvbe different, very low chance of getting the same random number

print("The Random Number without Seed:")
print(random.randint(0,100))
print(random.randint(0,100))
print(random.randint(0,100))
print(random.randint(0,100))
print(random.randint(0,100))

# RANDOM NUMBERS WITH SEED - if same seed then for all of the machine with the same seed will get the same set of random numbers
random.seed(101)
print("The Random Number with Seed:")
print(random.randint(0,100))
print(random.randint(0,100))
print(random.randint(0,100))
print(random.randint(0,100))
print(random.randint(0,100))
```

```python
# RANDOM WITH SEQUENCES

mylist = list(range(0,20))

print(mylist)

# Getting the random value from the list
print(f"The random value from the mylist: {random.choice(mylist)}")

# Getting the random values[set of value] from the list - values can bbe repeititive
print(f"List of the random values from the given list with repetitive values allowed:\n{random.choices(population= mylist, k=10)}")

# Getting the random values[set of value] from the list - values can't be repeititive
print(f"List of the random values from the given list with repetitive values allowed:\n{random.sample(population= mylist, k=10)}")
```

```python
#SHUFFLING THE LIST
print(f"The list before shuffle: {mylist}")

# in-place shuffling
random.shuffle(mylist)

print(f"The list after shuffling: {mylist}") 
```
